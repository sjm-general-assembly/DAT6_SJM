```python
import os
from pprint import pprint

import matplotlib.pyplot as plt
import pandas as pd
import twitter
import yaml
from pymongo import MongoClient
```

```python
credentials = yaml.load(open(os.path.expanduser('/Users/smarsh/Dropbox/ga/keys/api_cred.yml')))

auth = twitter.oauth.OAuth(credentials['ACCESS_TOKEN'],
                            credentials['ACCESS_TOKEN_SECRET'],
                            credentials['API_KEY'],
                            credentials['API_SECRET'])

twitter_api = twitter.Twitter(auth=auth)
```

```python
# set up twitter location IDs (from yahoo)
WORLD_WOE_ID = 1
US_WOE_ID = 23424977
CAN_WOE_ID = 23424775

# need to prefix ID param with an underscore (else twitter package)
# appends the ID value to the URL
world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
us_trends = twitter_api.trends.place(_id=US_WOE_ID)
can_trends = twitter_api.trends.place(_id=CAN_WOE_ID)

pprint(world_trends)
```

```python
# convert returned trends to sets
world_trends_set = set([trend['name'] for trends in world_trends
                        for trend in trends['trends']])

us_trends_set = set([trend['name'] for trends in us_trends
                     for trend in trends['trends']])
                     
can_trends_set = set([trend['name'] for trends in can_trends
                      for trend in trends['trends']])
```

```python
# generate a query string of all trends OR'd together
# time permitting, would error check against empty lists, and keeping maxlen < 1000
can_trends_list = list(can_trends_set)
all_can_trends_q_str = can_trends_list[0]
for trend in can_trends_list[1:]:
    all_can_trends_q_str += " OR " + trend
all_can_trends_q_str
```
