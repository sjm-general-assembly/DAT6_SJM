```python
import os
from pprint import pprint

import matplotlib.pyplot as plt
import pandas as pd
import twitter
import yaml
from pymongo import MongoClient
```

```python
credentials = yaml.load(open(os.path.expanduser('/Users/smarsh/Dropbox/ga/keys/api_cred.yml')))

auth = twitter.oauth.OAuth(credentials['ACCESS_TOKEN'],
                            credentials['ACCESS_TOKEN_SECRET'],
                            credentials['API_KEY'],
                            credentials['API_SECRET'])

twitter_api = twitter.Twitter(auth=auth)
```

```python
# set up twitter location IDs (from yahoo)
US_WOE_ID = 23424977
CAN_WOE_ID = 23424775

# need to prefix ID param with an underscore (else twitter package)
# appends the ID value to the URL
us_trends = twitter_api.trends.place(_id=US_WOE_ID)
can_trends = twitter_api.trends.place(_id=CAN_WOE_ID)

pprint(us_trends)
```

```python
pprint(can_trends)
```

```python
# convert returned trends to sets
us_trends_set = set([trend['name'] for trends in us_trends
                     for trend in trends['trends']])
                     
can_trends_set = set([trend['name'] for trends in can_trends
                      for trend in trends['trends']])
```

```python
us_trends_set
```

```python
can_trends_set
```

```python
# find which trends are common to both US and Canada
us_can_common_trends = us_trends_set.intersection(can_trends_set)
len(us_can_common_trends)
us_can_common_trends
```

```python
len(us_can_common_trends)
```

```python
# Find the trends for Canada only
can_only_trends_set = can_trends_set.difference(us_trends_set)
can_only_trends_set
```

```python
# Find the trends for US only
us_only_trends_set = us_trends_set.difference(can_trends_set)
us_only_trends_set
```

```python
# generate a query string of Canada only trends
can_only_trends_list = list(can_only_trends_set)
all_can_only_trends_q_str = '"' + can_only_trends_list[0] + '"'
for trend in can_only_trends_list[1:]:
    all_can_only_trends_q_str += ' OR "' + trend + '"'
all_can_only_trends_q_str
```

```python
# generate a query string of US only trends
us_only_trends_list = list(us_only_trends_set)
all_us_only_trends_q_str = '"' + us_only_trends_list[0] + '"'
for trend in us_only_trends_list[1:]:
    all_us_only_trends_q_str += ' OR "' + trend + '"'
all_us_only_trends_q_str
```

```python
# get the 100 US only trend tweets
us_search_results = twitter_api.search.tweets(q=all_us_only_trends_q_str, count=100)
us_search_results
```

```python
# get the 100 Canada only trend tweets
can_search_results = twitter_api.search.tweets(q=all_can_only_trends_q_str, count=100)
can_search_results
```

```python
# extract the just tweets from search results, pitch search metadata
can_tweets = can_search_results['statuses']
can_tweets
```

```python
# do the same for US (keep just tweets, pitch search metadata)
us_tweets = us_search_results['statuses']
us_tweets
```

```python
# get a connection to MongoDB, create a twitter db
c = MongoClient()
db = c.twitter
```

```python
db.canada.insert(can_tweets)
```

```python
db.us.insert(us_tweets)
```

```python
db.collection_names()
```

```python
us_most_followers = list(db.us.find(fields=['user.screen_name', 'user.followers_count']).sort('user.followers_count', -1).limit(5))
pd.DataFrame(us_most_followers).head()
```

```python
us_most_followers_df = pd.DataFrame({'followers_count': tweet['user'].get('followers_count'),
                                     'screen_name': tweet['user'].get('screen_name')}
                                    for tweet in us_most_followers)
us_most_followers_df.head()
```

```python
can_most_followers = list(db.canada.find(fields=['user.screen_name', 'user.followers_count']).sort('user.followers_count', -1).limit(5))
can_most_followers_df = pd.DataFrame({'followers_count': tweet['user'].get('followers_count'),
                                     'screen_name': tweet['user'].get('screen_name')}
                                    for tweet in can_most_followers)
can_most_followers_df.head()
```

```python
# find all retweeted tweets 
retweeted_us_tweets = db.us.find({'retweet_count': {'$gt': 0}})
retweeted_us_tweets.count()
```
